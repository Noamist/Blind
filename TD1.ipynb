```python
# --- Setup (pandas + XML) ---

# BASE: racine du dépôt en mode 'raw'
BASE = "https://raw.githubusercontent.com/sylvainhalle/8TRD151-TD1/main"

# Pandas est généralement déjà installé sur Colab, mais on l'importe explicitement
import pandas as pd

# Pour lire des XML avec pandas.read_xml, il faut lxml.
# On tente d'importer; si absent, on installe.
try:
    import lxml  # noqa: F401
except Exception:
    !pip -q install lxml
    import lxml  # noqa: F401

print("pandas:", pd.__version__)
print("BASE:", BASE)
```

```python
# @title Titre par défaut
# --- Helpers: lecture directe depuis GitHub raw ---

from urllib.parse import quote

def url(path: str) -> str:
    """Construit une URL raw à partir d'un chemin relatif au dépôt.
    Exemple: url('data/file.csv') -> https://.../main/data/file.csv
    """
    path = path.lstrip("/")
    # encode les espaces éventuels, etc.
    return f"{BASE}/{quote(path)}"

def read_csv(path: str, **kwargs) -> pd.DataFrame:
    """Lit un CSV depuis le dépôt GitHub raw."""
    return pd.read_csv(url(path), **kwargs)

def read_xml(path: str, **kwargs) -> pd.DataFrame:
    """Lit un XML depuis le dépôt GitHub raw.
    Selon la structure, vous devrez souvent préciser `xpath=...`.
    """
    return pd.read_xml(url(path), **kwargs)

# Mini-test (à adapter à vos vrais chemins):
# df = read_csv('data/exemple.csv')
# display(df.head())
# dx = read_xml('data/exemple.xml', xpath='.//item')
# display(dx.head())
```
